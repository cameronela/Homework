```python
# import dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

4. Pandas

```python
# (b-c) import Salaries.csv

df = pd.read_csv('Salaries.csv', index_col='playerID', skiprows=[1], header=0)
print(df.head(10))
```

```python
# (d) filter IDs

df_filtered = df[((df['teamID'] == 'ATL') | (df['teamID'] == 'HOU')) & (df['salary'] >= 1000000)]
print(list(df_filtered.index))
```

```python
# (e) use describe method

df_atl = df[df['teamID'] == 'ATL']
print(df_atl['salary'].describe())
```

```python
# (f) create a dictionary modeling the dataframe's structure

salaries_dict = {
    'playerID': [],
    'yearID': [],
    'teamID': [],
    'lgID': [],
    'salary': []}
for row in df.iterrows():
  salaries_dict['playerID'].append(row[0])
  for col in df.columns:
    salaries_dict[col].append(row[1][col])

print(salaries_dict['playerID'])
print(salaries_dict['yearID'])
print(salaries_dict['teamID'])
print(salaries_dict['lgID'])
print(salaries_dict['salary'])
```

```python
# (g) create dataframe from the ATL dataframe
df_new = pd.DataFrame(df_atl)
print(df_new.head(10))
new_names = {'yearID': 'a', 'teamID': 'b', 'lgID': 'c', 'salary': 'd'}
df_new = df_new.rename(columns=new_names)
print(df_new.head(10))
```

5. NumPy

```python
# (a) create a 2d array from a list

my_list = [[1, 2], [3, 4]]
my_arr = np.array(my_list)
print(my_list)
print(my_arr)
```

```python
# (b) examine array attributes

print(my_arr.ndim)
print(my_arr.shape)
print(my_arr.size)
print(my_arr.dtype)
print(my_arr.itemsize)
print(my_arr.data)
```

```python
# (c) use reshape and flatten
print(my_arr.reshape((4, -1)))
print(my_arr.flatten())
```

```python
# (d) slice an array

print(my_arr[1:])
print(my_arr[:1])
print(my_arr[2:4])
```

```python
# (e) use ndarray operations

print('argmin:', my_arr.argmin())
print('argmax:', my_arr.argmax())
print('min:', my_arr.min())
print('max:', my_arr.max())
print('mean:', my_arr.mean())
print('sum:', my_arr.sum())
print('std:', my_arr.std())
print('dot:', my_arr.dot([[2, 3], [3, 4]]))
print('square:', np.square(my_arr))
print('sqrt:', np.sqrt(my_arr))
print('abs:', np.abs([-1, -2]))
print('exp:', np.exp(my_arr))
print('sign:', np.sign([-1, 0, 1, -5]))
print('mod:', np.mod(my_arr, 4))
```

```python
# (f) examine numpy methods

print(np.arange(0, 10, 2))
print(np.ones(10))
print(np.zeros(10))
print(np.eye(10))
print(np.linspace(0, 5, 10))
print(np.concatenate(([1, 2], [3, 4])))
```

9. Matplotlib

```python
# (a) create a plot

x = np.linspace(0, 10, 11)
y = np.linspace(3, 7, 11)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y vs. x')
plt.grid()
plt.show()
```

```python
# (b) create multiple arrays

y1 = [2 * i + 3 for i in x]
y2 = [-3 * i + 5 for i in x]

plt.plot(x, y1, c='blue', label='y = 2x + 3')
plt.plot(x, y2, c='red', label='y = -3x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('lines')
plt.legend()
plt.show()
```

```python
# (c) make multiple plots

fig, ax = plt.subplots(1, 2, figsize=(16, 9))
ax[0].set_title('hello', fontdict={'fontsize': 10})
ax[1].set_title('world', fontdict={'fontsize': 20})
fig.show()
```

```python
# (e) limit axes and plot logarithmic axes

plt.xlim(1, 200)
plt.ylim(1, 200)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# (b) plot salaries dataframe with seaborn

sns.lineplot(df.iloc[:10, 3])
```
